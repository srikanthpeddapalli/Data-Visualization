# importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# set the style

```python
sns.set(style="darkgrid")
```

```python
sns.get_dataset_names()
```

# load dataset from the seaborn

```python
tips=sns.load_dataset("tips")
```

```python
tips.head()
```

```python
planets=sns.load_dataset("planets")
```

```python
planets.head()
```

```python
sns.relplot(x="size",y="tip",data=tips)
```

# now using the matplotlib

```python
planets.head()
```

```python
plt.scatter(x="mass",y="distance",data=planets)
```

# now ploting the line plot

```python
dots=sns.load_dataset("dots")
```

```python
dots.head()
```

```python
plt.scatter(x="time",y="firing_rate",data=dots)
```

# now using the seaborn

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri.head()
```
